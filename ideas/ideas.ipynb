## Feature Engineering

### EXT_SOURCES Ratio

- Correlation between EXT_SOURCE_1 and EXT_SOURCE_2: 0.21344587477440544
- Correlation between EXT_SOURCE_1 and EXT_SOURCE_3: 0.1873243239063459
- Correlation between EXT_SOURCE_2 and EXT_SOURCE_3: 0.10959492675500972

given the low correlation and high importance of the 3 external sources, a ratio feature could prove to be useful

### Important Features Clustering

A clustering based on the most important features i.e. EXT_SOURCES and C_to_Annuity could extract more information from the data. The feature importance data also aligns with the findings found on https://www.kaggle.com/competitions/home-credit-default-risk/discussion/64821, which creates a 500 neighbours target mean feature.
